# **XXZ Model Magnon Analysis**

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import cm
import mpmath as mp
import numpy as np
import scipy as scp
import sympy as sp
```

### Dispersion relations
**Without driving**
$$\varepsilon (\mathbf{k}) = h + 2dJS(\Delta - \gamma_{\mathbf{k}})$$
$$ \gamma_{\mathbf{k}} \equiv \frac{1}{d} \sum_i \cos k^i$$

## Data

```python
# Scalar parameters
h = 1
J = 1
S = 1
Delta = 2
d = 2

# Wavevectors
kx = np.linspace(-np.pi, np.pi, 100)
ky = np.copy(kx)
kx, ky = np.meshgrid(kx, ky)
def gamma(kx, ky):
    return (1/d) * (np.cos(kx) + np.cos(ky))
E = h + 2*d*J*S*(Delta - gamma(kx, ky))

# File name for the plot
pdf_name = "dispersion.pdf"
```

## Plotting

### 2D Plots

```python
%matplotlib inline
fig, ax = plt.subplots()
call = ax.contourf(kx, ky, E, levels=10, cmap=cm.viridis)
lines = ax.contour(kx, ky, E, levels=10, colors='black', linewidths=0.25)
colorbar = fig.colorbar(call, ax=ax)
colorbar.set_label(r'$\varepsilon(\mathbf{k})$')
ax.set(xlim=(-kx.max(), kx.max()),
       ylim=(-ky.max(), ky.max()),
       xlabel=r'$k_x$',
       ylabel=r'$k_y$',
       title='Magnon Spectrum of the 2D XXZ model')
fig.savefig("dispersion_contour.pdf")
```

### 3D Plots

```python
%matplotlib inline
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(kx, ky, E, cmap=cm.viridis)

ax.set(title='Magnon Spectrum of the 2D XXZ model',
       xlabel=r'$k_x$',
       ylabel=r'$k_y$',
       zlabel=r'$\varepsilon(\mathbf{k})$')
fig.savefig("dispersion_surface.pdf")
```
